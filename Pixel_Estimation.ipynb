```
import matplotlib.pyplot as plt
import random
import numpy as np
# from PIL import Image
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Reshape
from keras.layers import Conv2D, MaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras import backend as K
from keras.layers import Activation

import keras
from keras.callbacks import TensorBoard
import tensorflow as tf
from keras.callbacks import ModelCheckpoint
from keras import regularizers

def unpickle(pickled):
    # This function unpickles tha files into dictionaries / arrays.

    import pickle
    with open(pickled, 'rb') as fo:
        file = pickle.load(fo)
        fo.close()
    return file


def plot_image(an_image):
    plt.figure()
    plt.imshow(an_image)
    plt.colorbar()
    plt.show(block=True)


# TODO completing the images in the train set to find min-max values
def image_completion(image_set):
    array = np.array(image_set['x_data'])
    cnt = 0
    for i in range(0, array.shape[0]):
        array[i][15][15] = image_set['y_data'][i]
        cnt+=1
    return array

  
# def root_mean_squared_error(y_true, y_pred):
#     return np.abs(K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1)))

  
def data_augmentation(image_array, augment_size):
    '''this function takes the original full_image_array, and separates it into different image parts.
    original full_image_array size is (10000, 31, 31, 1).
    the output of the function is a  new, longer x_train, y_train, made by smaller images from the train set. '''

    from sklearn.feature_extraction import image
    patches = image.extract_patches_2d(image_array[0], (augment_size, augment_size))
    for index in range(1, image_array.shape[0]):
        patch = image.extract_patches_2d(image_array[index], (augment_size, augment_size), max_patches = 0.05, random_state = 42)
        patches = np.concatenate((patches, patch), axis=0)
    #creating new x_train, y_train
    x_new_train = patches
    center = int((augmentation_size-1)/2)
    y_train = np.array((x_new_train[0][center][center]))
    x_new_train[0][center][center] = 0

    for index in range(1, x_new_train.shape[0]):
        y_train = np.append(y_train, x_new_train[index][center][center])
        x_new_train[index][center][center] = 0

    y_train = np.expand_dims(y_train, 1)
    x_new_train = x_new_train.reshape(x_new_train.shape[0], augmentation_size, augmentation_size, 1)
    return x_new_train, y_train


def smoothL1(y_true, y_pred):
    HUBER_DELTA = 0.5
    x = K.abs(y_true - y_pred)
    if K._BACKEND == 'tensorflow':
        import tensorflow as tf
        x = tf.where(x < HUBER_DELTA, 0.5 * x ** 2, HUBER_DELTA * (x - 0.5 * HUBER_DELTA))
        return K.sum(x)

      
def data_normalization(x, min, max):
    """
    Normalize a list of sample image data in the range of 0 to 1
    : x: List of image data.  The image shape is (31, 31, 3)
    : return: Numpy array of normalized data
    """
    #[-1, 1] normalization (did work, but the return back did not)
    #return np.array(2*((x - min) / (max - min)) -1)
    
    #[0,1] normalization:
    return np.array((x - min) / (max - min))


def data_set():
    
  val_set1 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver9.pickle')
  val_set2 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver10.pickle')
  val_set = {'x_data': [], 'y_data':[]}

  val_set['x_data'].extend(val_set1['x_data'])
  val_set['y_data'].extend(val_set1['y_data'])
  val_set['x_data'].extend(val_set2['x_data'])
  val_set['y_data'].extend(val_set2['y_data'])

  val_set['x_data'] = np.array(val_set['x_data'])
  val_set['y_data'] = np.array(val_set['y_data'])

  test_set1 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver7.pickle')
  test_set2 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver8.pickle')
  test_set = {'x_data': [], 'y_data':[]}

  test_set['x_data'].extend(test_set1['x_data'])
  test_set['y_data'].extend(test_set1['y_data'])
  test_set['x_data'].extend(test_set2['x_data'])
  test_set['y_data'].extend(test_set2['y_data'])

  test_set['x_data'] = np.array(test_set['x_data'])
  test_set['y_data'] = np.array(test_set['y_data'])


  
  # Training set includes a dict with x-data and the y-data
  train_set1 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver1.pickle')
  train_set2 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver2.pickle')
  train_set3 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver3.pickle')
  train_set4 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver4.pickle')
  train_set5 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver5.pickle')
  train_set6 = unpickle('drive/elisra_dl_course_project/train data/train_data_1000_31X31_ver6.pickle')
  train_set = {'x_data': [], 'y_data':[]}
  train_set['x_data'].extend(train_set1['x_data'])
  train_set['y_data'].extend(train_set1['y_data'])
  train_set['x_data'].extend(train_set2['x_data'])
  train_set['y_data'].extend(train_set2['y_data'])
  train_set['x_data'].extend(train_set3['x_data'])
  train_set['y_data'].extend(train_set3['y_data'])
  train_set['x_data'].extend(train_set4['x_data'])
  train_set['y_data'].extend(train_set4['y_data'])
  train_set['x_data'].extend(train_set5['x_data'])
  train_set['y_data'].extend(train_set5['y_data'])
  train_set['x_data'].extend(train_set6['x_data'])
  train_set['y_data'].extend(train_set6['y_data'])

  train_set['x_data'] = np.array(train_set['x_data'])
  train_set['y_data'] = np.array(train_set['y_data'])

  del(train_set1, train_set2, train_set3, train_set4, train_set5, train_set6, val_set1, val_set2, test_set1, test_set2)
  return train_set, test_set, val_set


(train_set, test_set, val_set) = data_set()

augmentation_size = 11
input_size = (augmentation_size, augmentation_size, 1)
# input_size = augmentation_size*augmentation_size
img_x, img_y = augmentation_size, augmentation_size

batch_size = 2000
epochs =300
maxval = 2736 
minval = 487 


def input_handling_and_saving():
  
  full_image_array = image_completion(train_set)
  full_image_array = full_image_array.astype(int)
  
  #TrainSet bad data deletion:
  indexes = np.argwhere(full_image_array==0)
  index = []
  for i in range(len(indexes)):
    index.append(indexes[i][0])
  #index - is the rows index to remove from the full_image_arrays.
  np.argwhere(full_image_array[index]==0)
  full_image_array = np.delete(full_image_array, index, axis=0)

  full_val_array = image_completion(val_set)
  full_val_array = full_val_array.astype(int)

  full_test_array = image_completion(test_set)
  full_test_array = full_test_array.astype(int)

  #min-max value
  
  print(full_image_array.shape)
  print(full_test_array.shape)
  print(full_val_array.shape)
      
  norm_x = data_normalization(full_image_array, minval, maxval)
  (x_train, y_train) = data_augmentation(norm_x, augmentation_size)
  x_train = np.squeeze(np.reshape(x_train, (x_train.shape[0], img_y*img_x)))
  
  test_norm_x = data_normalization(full_test_array, minval, maxval)
  (new_x_test, new_y_test) = data_augmentation(test_norm_x, augmentation_size)
  new_x_test = np.squeeze(np.reshape(new_x_test, (new_x_test.shape[0], img_y*img_x)))
  
  val_norm_x = data_normalization(full_val_array, minval, maxval)
  (new_x_val, new_y_val) = data_augmentation(val_norm_x, augmentation_size)
  new_x_val = np.squeeze(np.reshape(new_x_val, (new_x_val.shape[0], img_y*img_x)))

  np.save('drive/elisra_dl_course_project/x_train_aug', x_train)
  np.save('drive/elisra_dl_course_project/y_train_aug', y_train)
  
  np.save('drive/elisra_dl_course_project/x_val_aug', new_x_val)
  np.save('drive/elisra_dl_course_project/y_val_aug', new_y_val)

  np.save('drive/elisra_dl_course_project/x_test_aug', new_x_test)
  np.save('drive/elisra_dl_course_project/y_test_aug', new_y_test)
  return x_train, y_train, new_x_val, new_y_val, new_x_test, new_y_test 

  

def input_loading(): 
  x_train = np.load('drive/elisra_dl_course_project/x_train_aug.npy')
  y_train = np.load('drive/elisra_dl_course_project/y_train_aug.npy')

  new_x_val = np.load('drive/elisra_dl_course_project/x_test_aug.npy')
  new_y_val = np.load('drive/elisra_dl_course_project/y_test_aug.npy')

  new_x_test = np.load('drive/elisra_dl_course_project/x_val_aug.npy')
  new_y_test = np.load('drive/elisra_dl_course_project/y_val_aug.npy')

  return x_train, y_train, new_x_val, new_y_val, new_x_test, new_y_test 



# (x_train, y_train, new_x_val, new_y_val, new_x_test, new_y_test) = input_handling_and_saving()
  
(x_train, y_train, new_x_val, new_y_val, new_x_test, new_y_test) = input_loading()



########################################
# FC method: these lines convert the input data to 4d numpy image.
# x_train = np.reshape(x_train, (x_train.shape[0], augmentation_size*augmentation_size))
# new_x_test = np.reshape(new_x_test, (new_x_test.shape[0], augmentation_size*augmentation_size))
# new_x_val = np.reshape(new_x_val, (new_x_val.shape[0], augmentation_size*augmentation_size))
# ########################################

########################################
#CNN -> FC method: these lines convert the input data to 4d numpy image.
x_train = np.reshape(x_train, (x_train.shape[0], augmentation_size, augmentation_size, 1))
new_x_test = np.reshape(new_x_test, (new_x_test.shape[0], augmentation_size, augmentation_size, 1))
new_x_val = np.reshape(new_x_val, (new_x_val.shape[0], augmentation_size, augmentation_size, 1))
########################################


print('train shape:', x_train.shape, y_train.shape)
print(x_train.shape[0], 'train samples')
print('test shape', new_x_test.shape, new_y_test.shape)
print(new_x_test.shape[0], 'test samples')
print('New "unseen" Validation shape', new_x_val.shape, new_y_val.shape)
print(new_x_val.shape[0], 'test samples')



# plot_image(np.reshape(x_train[random.randrange(1,500)], (augmentation_size,augmentation_size)))
# plot_image(np.reshape(x_test[random.randrange(1,500)], (augmentation_size,augmentation_size)))
# plot_image(np.reshape(x_val[random.randrange(1,500)], (augmentation_size,augmentation_size)))

from keras.layers.advanced_activations import LeakyReLU

model = Sequential()
model.add(Conv2D(32, (5, 5), strides = (1,1), padding='same', activation='relu', input_shape = input_size))
model.add(Conv2D(32, (5, 5), strides = (1,1), padding='same', activation='relu'))

model.add(Flatten())
model.add(Dense(128, kernel_initializer='normal',kernel_regularizer=regularizers.l2(0.000005),activation = 'relu'))
model.add(Dense(128, kernel_initializer='normal',kernel_regularizer=regularizers.l2(0.000005),activation = 'relu'))
model.add(Dense(1, activation='linear'))

print(model.summary())

filepath='drive/elisra_dl_course_project/CNNv16.weights-improvement-{epoch:02d}-{val_loss:.10f}.h5'

mc = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='auto', save_weights_only=True)
tbCallBack = TensorBoard(log_dir='./log', histogram_freq=1,write_graph=True,write_grads=True,batch_size=batch_size, write_images=True)

model.compile(optimizer='adam', loss='mean_squared_logarithmic_error', metrics=['mae'])

history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, verbose=1,
                    validation_data=(new_x_val, new_y_val),
                    shuffle=True,
                    callbacks=[mc])#, tbCallBack])

score = model.evaluate(new_x_test, new_y_test, verbose=1)


print('Test loss:', score)

# Loss Curves
plt.figure(figsize=[8, 6])
plt.plot(history.history['loss'], 'r', linewidth=3.0)
plt.plot(history.history['val_loss'], 'b', linewidth=3.0)
plt.legend(['Training loss', 'Validation Loss'], fontsize=18)
plt.xlabel('Epochs ', fontsize=16)
plt.ylabel('rmse Loss', fontsize=16)
plt.title('Loss Curves', fontsize=16)
plt.show(block=True)


plt.figure(figsize=[8, 6])
plt.plot(history.history['mean_absolute_error'], 'r', linewidth=3.0)
plt.plot(history.history['val_mean_absolute_error'], 'b', linewidth=3.0)
plt.legend(['Training MAE', 'Validation MAE'], fontsize=18)
plt.xlabel('Epochs ', fontsize=16)
plt.ylabel('MAE', fontsize=16)
plt.title('MAE', fontsize=16)
plt.show(block=True)

from keras.models import load_model

model.save('drive/elisra_dl_course_project/CNNv16.h5')  # creates a HDF5 file 'my_model.h5'

```
